### Set Multiple Values in DataFrame

````python
# we want to change every occurence of 'Sean Connery' to 'Sir Sean Connery'
#as earlier we did for a specific row only
# how can we do for all occurences of 'Sean Connery'
bond_5['Actor']=='Sean Connery'
#we can get boolean series
````

#### WRONG WAY OF DOING THIS

````python
actor_is_sean_connery=bond_5['Actor']=='Sean Connery'
````

````python
bond_5[actor_is_sean_connery]
````

````python
bond_5[actor_is_sean_connery]['Actor']='Sir Sean Connery'
#issue here is whenever we extract subset using square bracket
# we are creating a brand new dataframe 
#we are taking original rows from Bond_5 and pulling them out to a copy
#and we're storing copy elsewhere
#and of course we dont have even have it assigned to a variable ,it is completely lost
#But the original values in the 'Actor' column in Bond_5 have not been affected
# as we're targetting a separate dataframe, a separate subset which is a copy of original dataframe which has less rows
````

````python
bond_5
````

#### RIGHT WAY OF DOING

````python
# as suggested by warning correct way is to use .loc[]
bond_5.loc[actor_is_sean_connery,'Actor']
#this is reference to original dataframe
#we're not creating copy we're extracting original rows
#as we are using .loc we're connected to original DF
````

````python
bond_5.loc[actor_is_sean_connery,'Actor']='Sir Sean Connery'
````

````python
bond_5
#Whenever we use Bond_5[] We are extracting a copy, a separate dataframe
#We are extracting a slice that is separate object in computer's memory
# Whenever we use Bond_5.loc[] we're referencing existing row or rows in our original dataframe
````

````python
bond_8['Box Office']=bond_8['Box Office'].apply(convert_to_string_and_add_millions)
````

````python
bond_8.head(3)
````

## The ```.apply()``` method with Row Values

````python
bond_9=pd.read_csv('jamesbond.csv',index_col='Film')
bond_9.sort_index(inplace=True)
bond_9.head(3)
````

````python
#we want to create a separate column say review based on combination of row values
def good_movie(row): #row taken as list
    actor=row[1]
    budget=row[4]
    if actor=='Pierce Brosnan':
        return 'The best'
    elif actor=='Roger Moore' and budget>40:
        return 'Enjoyable'
    else:
        return 'I have no clue'
bond_9.apply(good_movie,axis='columns') #we need to pass row to the function
#we dont want to apply it to every value in column
#we want to apply horizantally i.e. row wise
````

````python
bond_9['Review']=bond_9.apply(good_movie,axis='columns')
````

````python
bond_9.head()
````

## The ```.copy()``` Method

````python
bond_10=pd.read_csv('jamesbond.csv',index_col='Film')
bond_10.sort_index(inplace=True)
bond_10.head(3)
````

````python
#copy method is the exact copy of an existing pandas object i.e. df or series
# but stores it completely separately in memory
bond_10['Director']
````

````python
directors=bond_10['Director']
directors.head(3)
````

````python
directors['A View to a Kill']='Mister John Glen'
````

````python
directors.head(3)
````

````python
bond_10.head(3)
#this modifies the original df 
# but what if we want to perform 15 diff opns but we dont want to keep overriding the original values
#and loose track of what we were originally storing
#we can create a copy of original df
````

````python
bond_10['Director'].copy() 
#brand new separated and isolated series
#and is not referenced to bond_10 df
````

````python
directors_=bond_10['Director'].copy() 
directors_.head(3)
````

````python
directors_['A View to a Kill']='Mister John Glen'
````

````python
directors_.head(3) # this obj or series modified
````

````python
bond_10.head(3) #original df not modified
````
